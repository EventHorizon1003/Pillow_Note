# Copying and Pasting Images onto Other Images

The copy() method will return a new image object with the same image as the Image object it was called on.

```python
copycatIm = catIm.copy()
print(copycatIm == catIm)
print(copycatIm)
print(catIm)
```

The paste() method will paste onto another image on top of it.

Syntax: "On what picture".paste("using what picture","coordinate")


```python
copycatIm.paste(im4,(40,40))
display(copycatIm)
```

```python
copycatIm.paste(im4,(80,80))
display(copycatIm)
```

Let do some funny things 

```python
sampleWidth , sampleHeight = catIm.size
print(str(sampleWidth) + " and " + str(sampleHeight))
copyWidth, copyHeight =  im4.size
print(str(copyWidth) + " and " + str(copyHeight))

samplecopytwo = catIm.copy()

for left in range(0,sampleWidth,copyWidth-300):
    for top in range(0,sampleHeight,copyHeight-300):
        print(left,top)
        samplecopytwo.paste(im4,(left,top))
```

# Resizing An Image

Using resize() methods to resize the image

Syntax: "Image_that_u_want".resize((width,height)) # the values must be in tuple // got two brackets

```python
# Let make the example of the picture be half

a , b = catIm.size

a = a/2
b = b/2

print(int(a))
copyIm = catIm.copy()
copyIm_resize = copyIm.resize((int(a),int(b)))
copyIm_resize.save("resize_pic.png")
```

```python
display(copyIm_resize)
```

# Rotating and Flipping Images

Using rotate() method to rotate the image (return a new image)

The second argument is for the expand option to enlarge the picture to fit the new image

Syntax: Img.rotate("degree",expand=T/F).save("name")


```python
copyIm_resize.rotate(90).save("90pic.jpg")
example = Image.open("90pic.jpg")
copyIm_resize.rotate(90,expand=True).save("90pic2.jpg")
example2 = Image.open("90pic2.jpg")

display(example)
display(example2)
```

Using transpose() method to get a mirror flip of an image

Syntax: Img.transpose(Image.FLIP_LEFT_RIGHT/FLIP_TOP_BOTTOM).save(name)


```python
copyIm.transpose(Image.FLIP_LEFT_RIGHT).save("flip1.jpg")
copyIm.transpose(Image.FLIP_TOP_BOTTOM).save("flip2.jpg")

flip1 = Image.open("flip1.jpg")
flip2 = Image.open("flip2.jpg")

display(flip1)
display(flip2)
```

# Changing Individual Pixels 

_getpixel()_ and _putpixel()_

Using getpixel() to get the color / RGBA 

-Syntax: Img.getpixel((coordinate))

Using putpixel() to change the RGBA

-Syntax: Img.putpixel((coordinate),(RGBA))

```python
im = Image.new("RGBA",(100,100))
#display(im) <= display the image
im.getpixel((50,50))

#Go throught the whole picture and examine the RGBA of all pixels
#for x in range(100):
#    for y in range(100):
#        print()

#Put the color onto it

from PIL import ImageColor # ImageColor get convert the string of color to value of RGBA
for x in range(100):
    for y in range(50,100):
        im.putpixel((x,y),ImageColor.getcolor('darkgrey',"RGBA"))        
display(im)

for x in range(0,100,5):
    for y in range(0,50):
        im.putpixel((x,y),ImageColor.getcolor("red","RGBA"))
display(im)
```
